#### plotting Multiple Lines
It's really easy to plot multiple lines in the same graph, just invoke the plt.plot function multiple times

```python
years = range(2000,2012)
apples = [0.895, 0.91,0.919, 0.926 , 0.929 , 0.931 , 0.934 , 0.936, 0.937, 0.9375 , 0.9372, 0.939] 
oranges = [0.962 , 0.941 , 0.930 , 0.918 , 0.908 , 0.907 , 0.904 ,0.901 , 0.898, 0.9 , 0.896, 0.9 ]
```

```python
plt.plot(years , apples)
plt.plot(years , oranges)
plt.xlabel("years")
plt.ylabel("yield(tons per hectare)")
```

#### chart Title and legend
To differentiate between multiple lines, we can include a legend within the graph using the plt.legend function. We also give the entire chart a title using the plt.title function.

```python
plt.plot(years , apples , marker="o")
plt.plot(years , oranges , marker="x")
plt.xlabel("years")
plt.ylabel("yield(tons per hectare)")
plt.title("Crop Yield in Kanto")
plt.legend(["Apples" , "oranges"])
```

#### styling line and marker
the plt.plot function supports many argument for styling lines and markers:
* color or c:set color of the line
* linestyle or ls: choose between a solid or dashed line
* linewidth or lw: set the width of the line.
* markersize or ms:set the size of the marker 
* markeredgecolor or mec: set the edge color for marker.
* markeredgewidth or mew: set the edge width for marker.
* markerfacecolor or mfc:set the fill color for markers.
* alpha:opacity of the plot

```python
plt.plot(years, apples, marker="s" , c="b" , ls="-" , lw=2 , ms=8 , mew=2, mec="navy")
plt.plot(years , oranges , marker="o" , c="r" , ls="--", lw=3 , ms=10 , alpha=.5)
plt.xlabel("year")
plt.ylabel("yield(tons per hectare)")
plt.title("crop Yields in Kanto")
plt.legend(['Apples'  , 'Oranges'])
```

the fmt argument provides a shorthand for specifying the line style, marker and line color, it can be provided as the third argument to plt.plot

fmt = '[marker][line][color]'

```python
plt.plot(years ,apples , 's-b' )# the third argument is fmt
plt.plot(years , oranges , 'o--r')

plt.xlabel('year')
plt.ylabel("yield(tons per hectare)")
plt.title("crop Yields in Kanto")
plt.legend(['Apples'  , 'Oranges']);
```

#### scatter plot
in a scatter plot, the values of 2 variable are plotted as points on a 2 dimensional grid.Additionally, you can also use a third variable to determine the size or color of the points.let's try out an example.


The [Iris flower dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set) provides sample measurments of sepals and petals for 3 species of flowers. The Iris dataset includes with the seaborn library and can be loaded as a Pandas dataframe.

```python
#Load data into a Pandas dataframe 
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

#### line chart using plt.plot .However the output is not very infomative as there are too many combination of the two properties within the dataset and thers doesnot seems to be simple relationship between them.

```python
plt.plot(flowers_df.sepal_length,flowers_df.sepal_width);
```

we can use a scatter plot to visualize how seplal width vary using the scatterplot function from seaborn (imported as sns).

```python
sns.scatterplot(flowers_df.sepal_length , flowers_df.sepal_width);
```

#### Adding Hues 
notice how the points in the above plot seem to from distinct cluster with some outliter. we can color the dot using the flower species as a hue. we can also make the points larger using the s argument.

```python
sns.scatterplot(flowers_df.sepal_length ,
                flowers_df.sepal_width ,
                hue=flowers_df.species,
                s=100);# s is size of each points
```

#### Customizing Seaborn Figure 
since Seaborn uses Matplotlib's plotting function internally, we can use function like plt.figure and plt.title to modify the figure.

```python
plt.figure(figsize=(13,6))
plt.title('sepal Dimension')

sns.scatterplot(flowers_df.sepal_length,
               flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100);
```

#### Plotting using Pandas Data Frames
seaborn has in-built support for Pandas data frames. insted of passing each column as a series, you can also pass column namees and use the data argument to pass the data frame.

```python
plt.title('sepal Dimenssions')
sns.scatterplot('sepal_length',
               'sepal_width',
               hue='species',
               s=100,
               data=flowers_df);
```

#### Histogram
A histogram represents the distribution of data by forming bins along the range of the data and then drawing bars to show the numbers of observation that fall in each bin.
As an example,Let's visualize the how the values of sepal width in the flowers dataset are diatribution.We can use plt.hist function to create a histogram.

```python
# load data into a Pandas dataframe
flowers_df = sns.load_dataset('iris')
```

```python
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
sns.set_style('darkgrid')
```

```python
plt.title('Distribution of sepal Width')
plt.hist(flowers_df.sepal_width);
```

#### Controlling the size and number of bin(interval # in mathematics)
control the number of bins, or the size of each bin using the bins argument.

```python
# Specifying the number of bin
plt.hist(flowers_df.sepal_width , bins=5);
```

```python
import numpy as np
np.arange(2,5,0.25)
```

```python
# specifying the boundaries of each bin 
plt.hist(flowers_df.sepal_width , bins=np.arange(2, 5 , 0.25));
```

```python
# Bins of unequal sizes 
plt.hist(flowers_df.sepal_width , bins=[1 , 3 , 4 , 4.5])
```

#### multiple Histogram 
similar to chart, we can draw multiple histogram in a single chart. We redue the opacity of each histogram , so the bars of one histogram don't hide the bars of others.



```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df =flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width , alpha=0.4 , bins=np.arange(2,5,0.25));
plt.hist(versicolor_df.sepal_width , alpha=0.4 , bins=np.arange(2 , 5,0.25));
plt.hist(virginica_df.sepal_width ,alpha=0.4 , bins=np.arange(2 , 5,0.25));
plt.legend(['setosa', 'versicolor' , 'virginica'])
```

stack multiple histogram on top of one another.

```python
plt.title('Distribution of sepal width')
plt.hist([setosa_df.sepal_width , versicolor_df.sepal_width , virginica_df.sepal_width],
        bins=np.arange(2,5,0.25),
        stacked=True)
plt.legend(['setosa', 'versicolor' , 'virginica'])
 
```

#### Bar chart 
Bar charts are quite similar to line chart i.e they show a sequence of values, however a bar is shown for each value, rather than points connected by lines. we can use the plt.bar function to draw a bar chart.

```python
years = range(2000 , 2006)
apples = [0.35 ,0.6 , 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9,  0.7, 0.6 ,0.8]
```

```python
plt.plot(years , apples);
```

```python
plt.bar(years , oranges);
```

```python
plt.bar(years , oranges)
plt.plot(years , oranges , 'o--r');
plt.title('Yields of oranges')
```

like  histogram, bars can also be stacked on top of one another . we use the bottom argument to plt.bar to achive this.

```python
plt.bar(years , apples)
plt.bar(years , oranges , bottom=apples);
plt.legend(['apples' , 'oranges'])
```

##### Bar plot with Average

we might want  to draw a bar chart to visualize how the average bill amount varies across different days of the week. one way to do this would be to compute the day-wise arrangement and then use plt.bar(try it as exercise).

```python
sns.barplot('day', 'total_bill' , data=tips_df);
```

The lines each bar represents the amount of variation in the values. For instance, it seems like the variation in the total bill was quite high on Friday, and lower on Saturday.

we can also specify a hue argument to compare bar plots side-by-side based on a third features e.g sex

```python
sns.barplot('day' , 'total_bill' , hue='sex' , data=tips_df);
```

#### Heatmap
A heatmap is used to visualize 2-dimentional data like a matrix or a table using colors.The best way to understand it is by looking at an example .we'll use another sample dataset from seaborn, called 'fights' , to visualize monthly passenger footfall at an airport over 12 years.

```python
df1 = sns.load_dataset('flights')
```

```python
df1
```

```python
 plt.plot(df1.passengers);
```

```python
flights_df = sns.load_dataset('flights').pivot("month" , "year" , "passengers")
```

```python
flights_df
```

flight_df is a matrix with one row each month and one column of each year. The values in the matrix show the number of passenger(in thousands) that visited the airport in a specific month of a specific year.We can use the sns.heatmap function to visualize the footfall at the airport.

```python
plt.title('No of passenger(1000s)')
sns.heatmap(flights_df);
```

```python
# we can also display the actual values in each block by specifying annnot=True , and use the cmap argument
# to change the color palette.
plt.title('No of passenger(1000s)')
sns.heatmap(flights_df , fmt='d' , annot=True , cmap='Blues');
```

#### plotting multiple charts in a grid 
Matplotlib and Seaborn also support plotting multiple charts in a grid, using plt.subplots , which returns a set of axes that can be used for plotting.

Here's a single grid showing the different type of charts we've covered in this tutorial.

```python
fig , axes = plt.subplots(2, 3)
axes[0,0].plot(years , oranges , 'o--r')
plt.tight_layout(pad=2)
```

```python
axes
```

```python
fig, axes = plt.subplots(2, 3, figsize=(16 ,8))

# use the axes for plotting 
axes[0,0].plot(years , apples ,  's-b')
axes[0,0].plot(years , oranges , 'o--r')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Yields (tons per hectare)')
axes[0,0].legend(['Apples' , 'Oranges'])
axes[0,0].set_title('Crop Yields in Kanto')


# Pass the axes in seaborn 
axes[0,1].set_title('sepal Length vs. sepal width')
sns.scatterplot(flowers_df.sepal_length, 
               flowers_df.sepal_width,
               hue=flowers_df.species,
               s=100,
               ax=axes[0,1]);
# use the axes for plotting 
axes[0,2].set_title('Distribution of sepal Width')
axes[0,2].hist([setosa_df.sepal_width , versicolor_df.sepal_width , virginica_df.sepal_width],
              bins=np.arange(2,5,0.25),
              stacked=True);
axes[0,2].legend(['setosa' , ' versicolor' , 'vieginica']);
# pass the axes into seaborn
axes[1,0].set_title('Restaurants Bill')
sns.barplot('day' , 'total_bill' , hue='sex' , data=tips_df , ax=axes[1,0]);

# Pass the axes into seaborn
axes[1,1].set_title('Flights traffic')
sns.heatmap(flights_df , cmap='Blues' , ax=axes[1,1]);

# plot an image using the axes
axes[1,2].set_title('Data science Meme')
axes[1,2].imshow(img)
axes[1,2].grid(False)
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.tight_layout(pad=2);
```
